```python
Data = Data[pd.notnull(Data['gender'])]
```

```python
print(Data['gender'].unique()) # 查看Sex特征有哪些值

# loc定位到目标行，对gender特征进行独热编码
Data.loc[Data['gender'] == 'M', 'gender'] = 0 # 另Sex等于male那行的Sex值为0
Data.loc[Data['gender'] == 'F', 'gender'] = 1 # 另Sex等于female那行的Sex值为1
```
